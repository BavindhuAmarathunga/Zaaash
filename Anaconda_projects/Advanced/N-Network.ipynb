```python
import pathlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
print(tf.__version__)
import tensorflow_docs as tfdocs
import tensorflow_docs.plots
import tensorflow_docs.modeling
```

```python
dataset=pd.read_csv("Example.csv")
```

```python
dataset = pd.get_dummies(dataset, prefix='', prefix_sep='')
dataset.tail()
```

```python
train_dataset = dataset.sample(frac=0.8,random_state=0)
test_dataset = dataset.drop(train_dataset.index)
```

```python
sns.pairplot(train_dataset[["Bugget", "H_Grade", "H_District", "Guest_Sum", "H_Accomadation", "H_Type", "W_DaN", "W_Theme"]], diag_kind="kde")
```

```python
train_stats = train_dataset.describe()
train_stats.pop("Bugget")
train_stats = train_stats.transpose()
train_stats
```

```python
train_labels = train_dataset.pop('Bugget')

test_labels = test_dataset.pop('Bugget')
```

```python
def norm(x):
  return (x - train_stats['mean']) / train_stats['std']
normed_train_data = norm(train_dataset)
normed_test_data = norm(test_dataset)
```

```python
def build_model():
  model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=[len(train_dataset.keys())]),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)
  ])

  optimizer = tf.keras.optimizers.RMSprop(0.001)

  model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
  return model
```

```python
model = build_model()
```

```python
model.summary()
```

```python
example_batch = normed_train_data[:10]
example_result = model.predict(example_batch)
example_result
```

```python
EPOCHS = 1000

history = model.fit(
  normed_train_data, train_labels,
  epochs=EPOCHS, validation_split = 0.2, verbose=0,
  callbacks=[tfdocs.modeling.EpochDots()])
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
plotter = tfdocs.plots.HistoryPlotter(smoothing_std=2)
```

```python
plotter.plot({'Basic': history}, metric = "mae")
plt.ylim([0, 5000])
plt.ylabel('budget')
```

```python
model = build_model()

# The patience parameter is the amount of epochs to check for improvement
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

early_history = model.fit(normed_train_data, train_labels, 
                    epochs=EPOCHS, validation_split = 0.2, verbose=0, 
                    callbacks=[early_stop, tfdocs.modeling.EpochDots()])
model.summary()
```

```python
plotter.plot({'Early Stopping': early_history}, metric = "mae")
plt.ylim([0, 5000])
plt.ylabel('budget')
```

```python
loss, mae, mse = model.evaluate(normed_test_data, test_labels, verbose=2)

print("Testing set Mean Abs Error: {:5.2f} Budget".format(mae))
```

```python
test_predictions = model.predict(normed_test_data).flatten()
a = plt.axes(aspect='equal')
plt.scatter(test_labels, test_predictions)
plt.xlabel('True Values [budget]')
plt.ylabel('Predictions [budget]')
lims = [0, 20000]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims, lims)
```

```python
error = test_predictions - test_labels
plt.hist(error, bins = 25)
plt.xlabel("Prediction Error [budget]")
_ = plt.ylabel("Count")
```
